```python
import pandas as pd
```

```python
BL1 = pd.read_csv("data.csv")
BL1
```

```python
BL1.drop(columns=['Unnamed: 0'], inplace = True)
```

```python
bl_clean = BL1.dropna(axis=0)
bl_clean
```

```python
y = bl_clean.median_house_value

filtered_features = ['longitude', 'latitude', 'housing_median_age', 'total_rooms','median_income']

X = bl_clean[filtered_features]

from sklearn.tree import DecisionTreeRegressor
model_1 = DecisionTreeRegressor()
model_1.fit(X,y)
```

```python
from sklearn.metrics import mean_absolute_error

predict_1 = model_1.predict(X)
mean_absolute_error(y, predict_1)
```

```python
from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
# Define model and chose test size (like test_size = .3)
model_2 = DecisionTreeRegressor()
# Fit model
model_2.fit(train_X, train_y)

# get predicted prices on validation data
val_predictions = model_2.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```
