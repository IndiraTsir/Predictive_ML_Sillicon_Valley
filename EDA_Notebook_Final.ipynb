# Cleaning my dataset

```python
# Search for the duplicates

df_silly_no_duplicates = df_silly.drop_duplicates()
df_silly_no_duplicates.info()

'''At the start we know we had 16512 rows Ã— 11 columns
it's the same values'''
```

```python
# Remove the ID column because useless for us

df_silly_no_duplicates.drop(columns=['Unnamed: 0'], inplace = True)
```

```python
df_silly_no_duplicates
```

```python
'''We have the choice to remove or to keep the attribute with missing values
Since we don't know how usefull it can be let's split them in 2 datasets
One dataset with the removed attribute
One dataset where the missing attribute is remplaced by 0
One dataset where it is remplaced by the mean'''
```

```python
df_silly_no_duplicates.info()
```

```python
# Replace the missing values for 0 

df_zero = df_silly_no_duplicates.fillna(0)
```

```python
# Remove the column with the missing values

df_no_bed = df_silly_no_duplicates.drop(columns=['total_bedrooms'])
```

```python
# Replace the column with the missing values for the median

df_median = df_silly_no_duplicates.fillna(df_silly_no_duplicates.median())
```

```python
# Replace the column with the missing values for the mean


df_mean = df_silly_no_duplicates.fillna(df_silly_no_duplicates.mean())
```

```python
# Let's see what kind of impact that have done on the dataset

df_zero.describe().T

# Original values: total_bedrooms	16336.0	539.319540	425.207704	2.0000	296.0000	435.0000	647.000000	6445.0000
```

## Treating the categorical values

```python
# Describe the values of the categorical attribute

df_silly_no_duplicates['ocean_proximity'].value_counts()
```

```python
# Replace values that are not strings

df_numerical = df_silly_no_duplicates.replace({'INLAND': 1, '<1H OCEAN': 2, 
                                 'NEAR OCEAN': 3, 'NEAR BAY': 4, 'ISLAND':5})
```

```python
df_numerical
```

```python
# Using a barplot let's see how ocean_proximity influence the median_house_value

sns.set(style="whitegrid")
sns.barplot(x="ocean_proximity", y="median_house_value", data = df_numerical)
```
